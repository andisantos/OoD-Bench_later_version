```python
import numpy as np
import random
random.seed(42)

mit_data = np.load("MIT_Indoors_samples_paths_and_labels_complete.npy")
labels, counts = np.unique(mit_data[:,1], return_counts=True)
mit_counts = dict(zip(labels, counts))
mit_counts
```

```python
len(mit_data)
```

```python
imagenet_arr = []
with open('/hadatasets/imagenet2012/filelist/train_filelist.txt', 'r') as f:
    line = f.readline()
    while line:
        x = line.split(" ")
        imagenet_arr.append(["/hadatasets/imagenet2012/train/images/"+ x[0], int(x[1][:-1])])
        line = f.readline()
print(imagenet_arr)
```

```python
imagenet_arr = np.asarray(imagenet_arr)
print(len(imagenet_arr))
print(imagenet_arr[12800])
labels, counts = np.unique(imagenet_arr[:,1], return_counts=True)
dict(zip(labels, counts))
```

```python
scene_to_imagenet_class_idx = {
    'bathroom': 9, #avestruz
    'bedroom': 479 , #car wheel
    'childs_room': 764, #rifle
    'classroom': 55, #grass snake
    'dressing_room': 0, #peixe
    'living_room': 130, #flamingo
    'studio': 162, #beagle
    'swimming_pool': 76, #tarantula
}
```

```python
dataset = []
for scene_class in scene_to_imagenet_class_idx.keys():
    imagenet_class = scene_to_imagenet_class_idx[scene_class]
    imagenet_filtered = np.where(imagenet_arr[:,1] == str(imagenet_class))
    
    selected_indexes = random.sample(list(imagenet_filtered[0]), k=mit_counts[scene_class])
    paths = imagenet_arr[selected_indexes]
    paths[:,1] = scene_class
    dataset.append(paths)
dataset = np.vstack(dataset)
len(dataset)

np.save("ImageNet8_samples_paths_and_labels.npy", np.asarray(dataset))
```
